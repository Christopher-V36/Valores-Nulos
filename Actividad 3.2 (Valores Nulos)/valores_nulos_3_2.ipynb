```python
#Librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carga de archivos y eliminación de las columnas no seleccionadas

```python
data = pd.read_csv("Tokio.csv")
data2 = data.copy()
data2 = data2.drop(columns=["id", "listing_url", "scrape_id", "source", "picture_url", "host_id", "host_url", "host_name","host_since","host_about","host_thumbnail_url","host_picture_url","host_neighbourhood","host_listings_count","host_total_listings_count","neighbourhood_group_cleansed","calendar_last_scraped","first_review","review_scores_accuracy","license","calculated_host_listings_count","calculated_host_listings_count_entire_homes","calculated_host_listings_count_private_rooms","calculated_host_listings_count_shared_rooms"])
data2
```

```python
#Convertir las columnas de tipo objeto que tenga valores numéricos a tipo numérico
#Primero se eliminan los caracteres especiales
data2["host_response_rate"] = data2["host_response_rate"].astype(str).str.replace('%', '')
data2["host_acceptance_rate"] = data2["host_acceptance_rate"].astype(str).str.replace('%', '')
#Se convierten a tipo numérico
data2["host_response_rate"] = pd.to_numeric(data2["host_response_rate"], errors='coerce')
data2["host_acceptance_rate"] = pd.to_numeric(data2["host_acceptance_rate"], errors='coerce')
```

```python
#Verificamos información del dataset
data2.info()
```

## Identificar valores nullos

```python
#Por columnas
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Por dataset
valores_nulos = data.isnull().sum().sum()
valores_nulos
```

## Sustitución de los valores nulos

```python
#Sustituimos los valores nulos de las columnas de tipo object por ? debido a que no se puede 
# hacer una interpolación de los datos 
data2["last_scraped"] = data2["last_scraped"].fillna("?")
data2["name"] = data2["name"].fillna("?")
data2["price"] = data2["price"].fillna("?")
data2["description"] = data2["description"].fillna("?")
data2["neighborhood_overview"] = data2["neighborhood_overview"].fillna("?")
data2["host_location"] = data2["host_location"].fillna("?")
data2["host_response_time"] = data2["host_response_time"].fillna("?")
data2["host_is_superhost"] = data2["host_is_superhost"].fillna("?")
data2["host_verifications"] = data2["host_verifications"].fillna("?")
data2["host_has_profile_pic"] = data2["host_has_profile_pic"].fillna("?")
data2["host_identity_verified"] = data2["host_identity_verified"].fillna("?")
data2["neighbourhood"] = data2["neighbourhood"].fillna("?")
data2["neighbourhood_cleansed"] = data2["neighbourhood_cleansed"].fillna("?")
data2["property_type"] = data2["property_type"].fillna("?")
data2["room_type"] = data2["room_type"].fillna("?")
data2["bathrooms_text"] = data2["bathrooms_text"].fillna("?")
data2["amenities"] = data2["amenities"].fillna("?")
data2["has_availability"] = data2["has_availability"].fillna("?")
data2["last_review"] = data2["last_review"].fillna("?")
data2["instant_bookable"] = data2["instant_bookable"].fillna("?")

#Corroboramos que no haya valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Sustituimos los valores nulos de las columnas de tipo numérico por la mediana de los 
# valores en las variables percentuales
data2["host_response_rate"] = data2["host_response_rate"].fillna(data2["host_response_rate"].median())
data2["host_acceptance_rate"] = data2["host_acceptance_rate"].fillna(data2["host_acceptance_rate"].median())
#Corroboramos que no haya valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Sustituimos la variable calendar_updated por un valor numérico concreto 0 debido a que la columna 
# esta llena de valores nulos
data2["calendar_updated"] = data2["calendar_updated"].fillna(0)
#Corroboramos que no haya valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Sustituimos las variables de review por la moda debido a que son variables categóricas
data2["review_scores_rating"] = data2["review_scores_rating"].fillna(data2["review_scores_rating"].mode()[0])
data2["review_scores_cleanliness"] = data2["review_scores_cleanliness"].fillna(data2["review_scores_cleanliness"].mode()[0])
data2["review_scores_checkin"] = data2["review_scores_checkin"].fillna(data2["review_scores_checkin"].mode()[0])
data2["review_scores_communication"] = data2["review_scores_communication"].fillna(data2["review_scores_communication"].mode()[0])
data2["review_scores_location"] = data2["review_scores_location"].fillna(data2["review_scores_location"].mode()[0])
data2["review_scores_value"] = data2["review_scores_value"].fillna(data2["review_scores_value"].mode()[0])
data2["reviews_per_month"] = data2["reviews_per_month"].fillna(data2["reviews_per_month"].mode()[0])
#Corroboramos que no haya valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#En el resto de variables numéricas se sustituyen los valores nulos por promedio debido a que es la 
# mejor opción para no alterar los datos y poder hacer una interpretación correcta de los mismos
data2["bathrooms"] = data2["bathrooms"].fillna(data2["bathrooms"].mean())
data2["bedrooms"] = data2["bedrooms"].fillna(data2["bedrooms"].mean())
data2["beds"] = data2["beds"].fillna(data2["beds"].mean())
data2["minimum_minimum_nights"] = data2["minimum_minimum_nights"].fillna(data2["minimum_minimum_nights"].mean())
data2["maximum_minimum_nights"] = data2["maximum_minimum_nights"].fillna(data2["maximum_minimum_nights"].mean())
data2["minimum_maximum_nights"] = data2["minimum_maximum_nights"].fillna(data2["minimum_maximum_nights"].mean())
data2["maximum_maximum_nights"] = data2["maximum_maximum_nights"].fillna(data2["maximum_maximum_nights"].mean())
data2["minimum_nights_avg_ntm"] = data2["minimum_nights_avg_ntm"].fillna(data2["minimum_nights_avg_ntm"].mean())
data2["maximum_nights_avg_ntm"] = data2["maximum_nights_avg_ntm"].fillna(data2["maximum_nights_avg_ntm"].mean())
#Corroboramos que no haya valores nulos
valores_nulos = data2.isnull().sum()
valores_nulos
```
